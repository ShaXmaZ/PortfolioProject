```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration we will create
 
#read in the data

df = pd.read_csv(r'C:\Users\Jazire\Downloads\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    #there is no null in data
```

```python
df.dtypes
```

```python

df['released'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
#dropping duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
#budget high correlation
#compony high correlation
#scatter plot with budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
#polt budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#starting to look ar correlation
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#look at compony
df.head()
```

```python
df_numberized = df

for col_name in df_numberized.columns:
    if(df_numberized[col_name].dtype == 'object'):
        df_numberized[col_name] = df_numberized[col_name].astype('category')
        df_numberized[col_name] = df_numberized[col_name].cat.codes
df_numberized.head()
```

```python
df
```

```python
correlation_matrix = df_numberized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numberized.corr()
```

```python
correlation_mat = df_numberized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
